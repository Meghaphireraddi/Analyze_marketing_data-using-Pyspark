```
df = spark.read.format("csv").option("header", "true").load("/FileStore/tables/Taskd.csv")
```

```
import pyspark
split_col = pyspark.sql.functions.split(df['"age;""job"";""marital"";""education"";""default"";""balance"";""housing"";""loan"";""contact"";""day"";""month"";""duration"";""campaign"";""pdays"";""previous"";""poutcome"";""y"""'], ';')
df = df.withColumn('age', split_col.getItem(0))
df = df.withColumn('job', split_col.getItem(1))
df = df.withColumn('marital', split_col.getItem(2))
df = df.withColumn('education', split_col.getItem(3))
df = df.withColumn('default', split_col.getItem(4))
df = df.withColumn('balance', split_col.getItem(5))
df = df.withColumn('housing', split_col.getItem(6))
df = df.withColumn('loan', split_col.getItem(7))
df = df.withColumn('contact', split_col.getItem(8))
df = df.withColumn('day', split_col.getItem(9))
df = df.withColumn('month', split_col.getItem(10))
df = df.withColumn('duration', split_col.getItem(11))
df = df.withColumn('campaign', split_col.getItem(12))
df = df.withColumn('pdays', split_col.getItem(13))
df = df.withColumn('previous', split_col.getItem(14))
df = df.withColumn('poutcome', split_col.getItem(15))
df = df.withColumn('y', split_col.getItem(16))
df = df.drop('"age;""job"";""marital"";""education"";""default"";""balance"";""housing"";""loan"";""contact"";""day"";""month"";""duration"";""campaign"";""pdays"";""previous"";""poutcome"";""y"""')
```

```
def fun(x):
  return x.replace('""',"")
def toc(x):
  return x.replace('"',"")

udf_ = udf(lambda z:fun(z))
udf1_ = udf(lambda z:toc(z))
df = df.withColumn('age',udf1_('age'))
df = df.withColumn('job',udf_('job'))
df = df.withColumn('marital',udf_('marital'))
df = df.withColumn('education',udf_('education'))
df = df.withColumn('default',udf_('default'))
df = df.withColumn('housing',udf_('housing'))
df = df.withColumn('loan',udf_('loan'))
df = df.withColumn('contact',udf_('contact'))
df = df.withColumn('month',udf_('month'))
df = df.withColumn('poutcome',udf_('poutcome'))
df = df.withColumn('y',udf_('y'))
df = df.withColumn('y',udf1_('y'))
```
